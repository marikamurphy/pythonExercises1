```python
import numpy as np
import pandas as pd
%matplotlib inline
t = np.linspace(0,10,200)
df = pd.DataFrame({
    'X': np.sin(t) + 0.1*np.random.randn(200),
    'Y': np.sin(2**t) + 0.1*np.random.randn(200)
})

df.abs()
```

```python
means= df.mean()

print(means)

def subtract_mean(x):
    return x-means

df.apply(subtract_mean, axis=1)
```

```python
sampledf = df.rolling(10).mean()
sampledf.plot()

```

```python
#Data Aggregation
gb = df.groupby('X')
gb.groups
for groupname, subdf in gb:
    print(groupname)
    print(subdf)
    
```

```python
df2 = df.reset_index()
df2
def even(x):
    return x%2==0
gb2 = df2.groupby(even)
gb2.groups
```
